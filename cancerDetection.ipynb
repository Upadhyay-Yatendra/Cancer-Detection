```python
import pandas as pd
import seaborn as sns
```

```python
data = pd.read_csv('data.csv')
data.head()
```

```python
data.describe()
```

```python
# Clean the data
sns.heatmap(data.isnull(), cbar=False)
```

```python
print(data.isna().sum())  # Shows NaN count per column
```

```python
data.drop(['Unnamed: 32' , "id"], axis=1, inplace=True)
```

```python
data.head()
```

```python
data.diagnosis = [1 if each == "M" else 0 for each in data.diagnosis]
```

```python
row_count = len(data.index)

print("Total rows:", row_count)
```

```python
data.head()
```

```python
data["diagnosis"] = data["diagnosis"].astype("category" , copy=False)
data["diagnosis"].value_counts().plot(kind='bar')
```

```python
# cleanign the data has been done, now we can start with the model
```

<!-- Now divind the dataset -->
divide into target variable and predictors

```python
y = data.diagnosis.values #our target variable
x = data.drop(["diagnosis"],axis=1)
```

```python
print(y)
```

```python
print(x)
```

```python
# Now we will normalise the data
from sklearn.preprocessing import StandardScaler
# create scaler object
sc = StandardScaler()
# fit the scaler to the feature data and transform
X_scaled = sc.fit_transform(x)
print(X_scaled)
```

```python
# split the data into training and testing sets
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size = 0.3, random_state = 0)
```

```python
# actually train the model now
from sklearn.linear_model import LogisticRegression
# create the lr model
lr = LogisticRegression()
# train the model
lr.fit(X_train, y_train)
```

```python
# predict the target variable for the test set
y_pred = lr.predict(X_test)
```

```python
# evaluate the model
from sklearn.metrics import accuracy_score

accuracy = accuracy_score(y_test, y_pred)
print(accuracy)
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test, y_pred))
```
